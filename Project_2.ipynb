Import Numpy, Pandas, matplotlib, Seaborn, and the raw data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv")
df.head()
df.sample(10)
```

Eaxam data quality for attributes and null values

```python
print('Shape (rows, columns):', df.shape)

print('\nColumn names:')
print(df.columns)

print('\nData types and non-null counts:')
df.info()

print('\nSummary statistics (numeric columns):')
df.describe()
```

Exam missing value for each column

```python
print('Missing values per column:')
print(df.isna().sum())

print('\nPercentage of missing values per column:')
print((df.isna().mean() * 100).round(2))

print('\nNumber of unique values per column:')
print(df.nunique())
```

Clean up column names and standardize format

```python
df.columns = (
    df.columns
      .str.strip()
      .str.upper()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)
df.columns
```

I pick "Race" as initial measurement

```python
df['RACE'].unique()[:20]
```

Clean up null value by replacing it with "nan"

```python
df['RACE'] = df['RACE'].replace('?', np.nan)
df['WEIGHT'] = df['WEIGHT'].replace('?', np.nan)
df['RACE'].unique()
print("Unique values in 'RACE' column:")
print(df['RACE'].unique())

print("\nUnique values in 'WEIGHT' column:")
print(df['WEIGHT'].unique())
```

Double check to make sure no null value for "Race"

```python
print(df['RACE'].value_counts())
```

Visulize the patients by race with a histogram chart. Caucasian is the majority of the data. Hence, I further drill down race and pick Caucasian to compare all other races for visit by medical specialty

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 4))

sns.histplot(df['RACE'], bins=30, kde=True, ax=axes)
axes.set_title('RACE')

plt.tight_layout()
plt.show()
```

Top 10 medical specialty visit by caucasian paitients - top 3 are Internal Medicine, Emergency Trauma, and Family General Practice

```python
df1 = df[df['RACE']=='Caucasian']
df2 = df1[df1['MEDICAL_SPECIALTY'] != '?']
df3 = df2['MEDICAL_SPECIALTY'].value_counts().nlargest(10).index.tolist()
df4 = df2[df2['MEDICAL_SPECIALTY'].isin(df3)].copy()

fig, axes = plt.subplots(1, 1, figsize=(20, 4))

sns.histplot(df4['MEDICAL_SPECIALTY'], bins=30, kde=True, ax=axes)
axes.set_title('Caucasian Paitient')

plt.tight_layout()
plt.show()
```

Top 10 medical specialty visit by non-caucasian paitients - top 3 are Internal Medicine, Family General Practice, and Emergency Trauma. The top 2 and 3 are reserse when comparing to caucasian paitients only.

```python
df1 = df[df['RACE'] != 'Caucasian']
df2 = df1[df1['MEDICAL_SPECIALTY'] != '?']
df3 = df2['MEDICAL_SPECIALTY'].value_counts().nlargest(10).index.tolist()
df4 = df2[df2['MEDICAL_SPECIALTY'].isin(df3)].copy()

fig, axes = plt.subplots(1, 1, figsize=(20, 4))

sns.histplot(df4['MEDICAL_SPECIALTY'], bins=30, kde=True, ax=axes)
axes.set_title('All Paitient')

plt.tight_layout()
plt.show()
```

Overall the most data are caucasian race. Top medical specialty visited by paitients are similar; However, the top 2 and 3 are different when comparing caucasian to non-caucasian paitients.
